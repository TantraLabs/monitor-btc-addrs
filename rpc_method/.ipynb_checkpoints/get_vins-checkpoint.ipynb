```python
from bitcoinrpc.authproxy import AuthServiceProxy
import pandas as pd
```

```python
rpc_connection = AuthServiceProxy("http://%s:%s@ec2-3-81-20-12.compute-1.amazonaws.com:8332"%("test", "test"))
```

```python
wallet_data = pd.read_csv('wallet_data.csv')
wallet_txs = pd.DataFrame()
progress = 0

for txid in wallet_data['txid']:
    # print('main tx', txid)
    raw_data = rpc_connection.getrawtransaction(txid, True)
    vins = raw_data['vin']
    for vin in vins:
        real_vin = rpc_connection.getrawtransaction(vin['txid'], True)['vout'][vin['vout']]
        tx_in = {}
        
        tx_in['timestamp'] = raw_data['time']
        tx_in['tx_hash']   = txid
        tx_in['amount']     = real_vin['value']
        tx_in['tx_type']   = 'deposit'
        tx_in['type']      = 'btc'
        tx_in['address']   = real_vin['scriptPubKey']['addresses'][0]
        wallet_txs = wallet_txs.append(pd.DataFrame(tx_in,index=[0]))
      
        # print('vin tx', vin['txid'])
        # print(real_vin['value'])
        # print(real_vin['scriptPubKey']['addresses'])
    
    progress+=1
    if (progress % 10 == 0):
        print(f"progress {progress}/{len(wallet_data['txid'])} txs")        
    

wallet_txs = wallet_txs.set_index('address')
wallet_txs['address'] = wallet_txs.index

wallet_txs
```
