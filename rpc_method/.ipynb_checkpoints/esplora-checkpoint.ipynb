```python
import pandas as pd
import requests
from collections import defaultdict
import time

input_df = pd.read_csv('addresses.csv', index_col=0)
input_df.head()
```

```python
btc_hashes = input_df['0'].values
addr_data = defaultdict(list)
i = 0
for btc_hash in btc_hashes[:100]: # limiting for testing
    last_tx = ''
    while(True):
        URL = "https://blockstream.info/api/address/{}/txs/chain"
        if last_tx != '':
            URL+=f'/{last_tx}'

        data = requests.get(URL.format(btc_hash))


        data = data.json()
        addr_data[btc_hash].extend(data)
        if (len(data) < 25):
            break
        last_tx = data[-1]['txid']
    i+=1
    if i % 20 == 0:
        print(f"{i}/{len(btc_hashes)}")
    # time.sleep(1)
print(f"{i}/{len(btc_hashes)}")
```

```python
history = pd.DataFrame()

for addr in addr_data.keys():
    for tx in addr_data[addr]:
        amount = 0
        for v_in in tx['vin']:
            if v_in['prevout']['scriptpubkey_address'] == addr:
                amount += v_in['prevout']['value']
        amount = -amount

        for v_out in tx['vout']:
            if v_out['scriptpubkey_address'] == addr:
                amount += v_out['value'] 
                
#         amount /= 1e8
        
        tx_history = {
            'time': tx['status']['block_time'],
            'addr': addr,
            'value': amount,
            'txid':tx['txid']
        }
        tx_df = pd.DataFrame(tx_history, index=[0])
        history = history.append(tx_df, ignore_index=True)
#     break

history
```

```python
history = history.sort_index()
history['balance']  =  history.value.cumsum().div(1e8)
```

```python
wallet_txs = pd.DataFrame()
progress = 0

for addr in addr_data.keys(): # here the addr_data is same that was calculated before
    for tx in addr_data[addr]:
        for vin in tx['vin']:
            tx_in = {}

            tx_in['timestamp'] = tx['status']['block_time']
            tx_in['tx_hash']   = tx['txid']
            tx_in['amount']    = vin['prevout']['value']
            tx_in['tx_type']   = 'deposit'
            tx_in['type']      = 'btc'
            tx_in['address']   = vin['prevout']['scriptpubkey_address']
            wallet_txs = wallet_txs.append(pd.DataFrame(tx_in,index=[0]))


    progress+=1
    if (progress % 10 == 0):
        print(f"progress {progress}/{len(addr_data)} txs")        
print(f"progress {progress}/{len(addr_data)} txs")        
    

wallet_txs = wallet_txs.set_index('address')
wallet_txs['address'] = wallet_txs.index

wallet_txs
```
