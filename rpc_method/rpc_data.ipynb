#### In BTC Node, I have already imported these addreeses under 'test' label to watch only addresses and rescanned the wallet for all the transactions.


```python
rpc_connection = AuthServiceProxy("http://%s:%s@ec2-3-81-20-12.compute-1.amazonaws.com:8332"%("test", "test"))
```

```python
labeled_addr = rpc_connection.getaddressesbylabel("test")
len(labeled_addr)
```

```python
df = pd.DataFrame()
count_val = 100
skip_val = 0
while(True):
    skip_val+=count_val
    get_txs = rpc_connection.listtransactions("test", count_val, skip_val, True) # change 'test' to '*' to include send tx as well
    if get_txs == []:
        break
    df = df.append(pd.DataFrame(get_txs), ignore_index=True)
    print(f"fetched {len(df)} txs")


df
```

```python
# df_by_addr = df.groupby('address')
# len(df_by_addr)
```

```python
txids = df.txid.values
tx_df = pd.DataFrame()
total_txs = len(txids)
i = 0
for txid in txids:
    raw_data = rpc_connection.gettransaction(txid, False, True)
    vout_addr = [", ".join(raw_data['decoded']['vout'][i]['scriptPubKey']['addresses']) for i in range (len(raw_data['decoded']['vout']))]
    vin_addr = [(raw_data['decoded']['vin'][i]['txid'], raw_data['decoded']['vin'][i]['vout']) for i in range(len(raw_data['decoded']['vin']))]
    line = [[raw_data['decoded']['txid'],vin_addr, vout_addr]]
    temp_df = pd.DataFrame(line, columns=['txid','vin_raw_list','vout_list'])
    tx_df = tx_df.append(temp_df, ignore_index = True)
    i+=1
    print(f"progress {i}/{total_txs} txs")
```

```python
merged_df = pd.merge(df, tx_df)
merged_df
```

```python
merged_df.to_csv('temp.csv')
```
